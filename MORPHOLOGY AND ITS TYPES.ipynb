# IMPORTING LIB


```python
import cv2 as cv
import matplotlib.pyplot as plt 
import numpy as np
```

```python
img=cv.imread("C:/Users/shiya/Downloads/noise2.png",0)
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
_,mask=cv.threshold(img,220,255,cv.THRESH_BINARY)
kernal=np.ones((3,3),np.uint8)
dilation=cv.dilate(mask,kernal,iterations=1)
erode=cv.erode(mask,kernal,iterations=1)
titles=['image','mask','dilate','erode']
images=[img,mask,dilation,erode]
for i in range(len(images)):
    plt.subplot(2,2,i+1),plt.imshow(images[i])
    plt.title(titles[i])
plt.show
```

# morphological opening-EROSIONfollowed by DILATION


```python
open =cv.morphologyEx(img,cv.MORPH_OPEN,kernal)
```

```python
cv.imshow("img",open)
cv.waitKey(2)
cv.destroyAllWindows()
```

# morphological closing-DILUTION followed by EROSION

```python
open1=cv.morphologyEx(img,cv.MORPH_CLOSE,kernal)
```

```python
cv.imshow("img1",open)
cv.imshow("img",open1)
cv.waitKey(0)
cv.destroyAllWindows()
```

# morphological gradient-erosion,dilation,difference of both

```python
open2=cv.morphologyEx(img,cv.MORPH_GRADIENT,kernal)
```

```python
cv.imshow("img1",open2)
cv.imshow("img",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

# TOPHAT-OPENING AND DIFF OF ORIGINAL AND OPENING

```python
open3=cv.morphologyEx(img,cv.MORPH_TOPHAT,kernal)
```

```python
cv.imshow("img1",open3)
cv.imshow("img",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

# BLACK HAT-OPOSSITE OF TOP HAT

```python
open4=cv.morphologyEx(img,cv.MORPH_BLACKHAT,kernal)
```

```python
cv.imshow("img1",open4)
cv.imshow("img",img)
cv.waitKey(0)
cv.destroyAllWindows()
```
